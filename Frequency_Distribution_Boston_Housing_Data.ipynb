# Using the Boston Housing Data for Frequency Distribution

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Getting the boston housing data from the scikitlearn datasets
from sklearn.datasets import load_boston
```

```python
# Creating an object 
boston = load_boston()
```

```python
boston.keys()
```

```python
# Description of the data set
print(boston['DESCR'])
```

```python
# Creating a dataframe for features variables
data = pd.DataFrame(boston['data'],columns=boston['feature_names'])

data.head()
```

```python
data['MEDV'] = boston['target']
```

```python
data.head()
```

So I have created the entire dataset I want to explore. The target column in the MEDV column that gives the median value of owner occupied homes.

```python
# Checking for null values
data.isna().sum()
```

```python
# Frequency distribution of CRIM:
crim_freq = pd.crosstab(index=data['CRIM'],columns='Frequency')
crim_freq
```

```python
# Graphical representation
sns.histplot(x='CRIM',data=data)
```
