# Wind Speed data: 

```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# Importing the dataset
sourcedata = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(sourcedata, sep = "\s+", parse_dates = [[0,1,2]]) 
data.head() # showing the head of the data
```

```python
# I find here that years are 2061, so the years must be corrected by decreasing the years by 100

# Creating a function to correct the years
def correct_years(y):
  year = y.year - 100 if y.year > 1989 else y.year
  return datetime.date(year, y.month, y.day)

# applying the function
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(correct_years)

# Looking at the data now
data.head()
```

```python
data.info()
```

```python
# Changing the date column Yr_Mo_Dy to data-type datetime64
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

# setting the correct date column 'Yr_Mo_Dy'
data = data.set_index('Yr_Mo_Dy')

data.head() # Looking at the data again
```

```python
data.info()
```

```python
# Trying to see the null-values in the data

# Summarizing the total number of missing values in the data
data.isna().sum()
```

```python
# Number of non-missing values
data.notnull().sum()
```

```python
# Dropping the null values since they have to be ignored
data1 = data.dropna()
```

```python
# Calculating the mean windspeeds of the windspeeds over all locations at all times
data1.mean().mean()
```

```python
# Statistics for each location at all times
loc_stats=data1.describe(percentiles=[])
loc_stats
```

```python
# Statistics for each day at all locations
day_stats = pd.DataFrame()

# All statistics:
day_stats['min'] = data1.min(axis = 1) # minimum
day_stats['max'] = data1.max(axis = 1) # maximum 
day_stats['mean'] = data1.mean(axis = 1) # mean
day_stats['std'] = data1.std(axis = 1) # standard deviations

day_stats.head()
```

```python
# average windspeed in January for each location
data1.loc[data1.index.month == 1].mean()
```

```python
# Downsampling the record to a yearly frequency for each location
data1.groupby(data1.index.to_period('A')).mean()
```

```python
# Downsampling the record to a monthly frequency for each location
data1.groupby(data1.index.to_period('M')).mean()
```

```python
# Downsample the record to a weekly frequency for each location
data1.groupby(data1.index.to_period('W')).mean()
```

```python
# Calculating the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week 
weekly = data1.resample('W').agg(['min','max','mean','std'])
weekly.loc[weekly.index[1:53], "RPT":"MAL"] .head(10)
```
