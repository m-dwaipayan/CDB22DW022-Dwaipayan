# Exploratory data analysis
Checking for the distribution of the different columns.
Since only three columns are continuous (average glucose level, age and bmi) we check for the distributions of them.

**Checking for average glucose level**

```python
sns.distplot(data['avg_glucose_level'])
```

```python
sns.histplot(data,x='avg_glucose_level',hue='gender',kde=True)
```

```python
sns.histplot(data,x='avg_glucose_level',hue='hypertension',kde=True)
```

```python
sns.histplot(data,x='avg_glucose_level',hue='heart_disease',kde=True)
```

```python
sns.histplot(data,x='avg_glucose_level',hue='ever_married',kde=True)
```

```python
sns.histplot(data,x='avg_glucose_level',hue='work_type',kde=True)
```

```python
sns.histplot(data,x='avg_glucose_level',hue='Residence_type',kde=True)
```

```python
sns.histplot(data,x='avg_glucose_level',hue='smoking_status',kde=True)
```

The KDE plot of avg_glucose_level is bimodal, and those for it along with hues follow similar pattern. Thus avg_glucose_level is not normally distributed.

**Checking for bmi**

```python
sns.distplot(data['bmi'])
```

```python
sns.kdeplot(data['bmi'])
```

```python
sns.histplot(data,x='bmi',hue='gender',kde=True)
```

```python
sns.histplot(data,x='bmi',hue= 'heart_disease',kde=True)
```

```python
sns.histplot(data,x='bmi',hue='work_type',kde=True)
```

```python
sns.histplot(data,x='bmi',hue='Residence_type',kde=True)
```

```python
sns.histplot(data,x='bmi',hue='smoking_status',kde=True)
```

Since neither the histogram nor the kde plot is bell shaped for bmi, the bmi data is not normally distributed.

**Checking for age**

```python
sns.distplot(data['age'])
```

Since the kde plot for age is also not bell-shaped, it is not normally distributed.

# Checking for Skewness

```python
sns.distplot(data['avg_glucose_level'],kde=True,label=True)
plt.axvline(x=data['avg_glucose_level'].median(),color='red',ls='--')
plt.axvline(x=data['avg_glucose_level'].mean(),color='green',ls='--')
```

```python
sns.distplot(data['bmi'],kde=True,label=True)
plt.axvline(x=data['bmi'].median(),color='red',ls='--')
plt.axvline(x=data['bmi'].mean(),color='green',ls='--')
```

```python
sns.distplot(data['age'],kde=True,label=True)
plt.axvline(x=data['age'].median(),color='red',ls='--')
plt.axvline(x=data['age'].mean(),color='green',ls='--')
```
